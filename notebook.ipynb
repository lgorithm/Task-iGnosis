```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore')
```

```python
transaction_df = pd.read_csv('transaction_data.csv')
purchase_behaviour_df = pd.read_csv('purchase_behaviour.csv')
transaction_df
```

```python
transaction_df = transaction_df.drop_duplicates()
print('No of duplicates: {}'.format(sum(transaction_df.duplicated())))
```

```python
# Identify the top 3 most profitable products based on total sales
top_products = (
    transaction_df.groupby(["PROD_NBR", "PROD_NAME"])["TOT_SALES"]
    .sum()
    .reset_index()
    .sort_values(by="TOT_SALES", ascending=False)
    .head(3)
)

top_products
```

```python
# Set style
sns.set_style("whitegrid")

# Plot top 3 most profitable products
plt.figure(figsize=(10, 5))
sns.barplot(
    x=top_products["TOT_SALES"],
    y=top_products["PROD_NAME"],
    palette="viridis"
)
plt.xlabel("Total Sales ($)")
plt.ylabel("Product Name")
plt.title("Top 3 Most Profitable Products")
plt.show()
```

```python
# Identify loyal customers based on total spending and purchase frequency
loyal_customers = (
    transaction_df.groupby("LYLTY_CARD_NBR")["TOT_SALES"]
    .sum()
    .reset_index()
    .sort_values(by="TOT_SALES", ascending=False)
)

# Merge with customer demographics
loyal_customers = loyal_customers.merge(purchase_behaviour_df, on="LYLTY_CARD_NBR", how="left")

# Group by LIFESTAGE and PREMIUM_CUSTOMER to analyze trends
customer_segments = (
    loyal_customers.groupby(["LIFESTAGE", "PREMIUM_CUSTOMER"])["TOT_SALES"]
    .sum()
    .reset_index()
    .sort_values(by="TOT_SALES", ascending=False)
)

customer_segments.head(3)
```

```python
# Plot top customer segments by total spending
plt.figure(figsize=(12, 6))
sns.barplot(
    x=customer_segments["TOT_SALES"],
    y=customer_segments["LIFESTAGE"] + " - " + customer_segments["PREMIUM_CUSTOMER"],
    palette="mako"
)
plt.xlabel("Total Sales ($)")
plt.ylabel("Customer Segment")
plt.title("Top Spending Customer Segments")
plt.show()
```
